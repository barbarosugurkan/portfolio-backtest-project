# Question 1
Find all the tracks that have a length of 5,000,000 milliseconds or more.

```python
%%sql
SELECT Name, Milliseconds
FROM Track
WHERE Milliseconds >= 5000000
```

# Question 2

Find all the invoices whose total is between $5 and $15 dollars.

```python
%%sql
SELECT InvoiceID, Total
FROM Invoice
WHERE Total BETWEEN 5 AND 15
LIMIT 10
```

# Question 3

Find all the customers from the following States: RJ, DF, AB, BC, CA, WA, NY.

```python
%%sql 
SELECT FirstName,LastName,State, Company
FROM Customer
WHERE State IN ("RJ", "DF", "AB", "BC", "CA", "WA", "NY")
```

# Question 4

Find all the invoices for customer 56 and 58 where the total was between $1.00 and $5.00.

```python
%%sql
SELECT InvoiceId, Total, CustomerId
FROM Invoice
WHERE (CustomerId = 56 OR CustomerId = 58) AND Total BETWEEN 1.00 AND 5.00
```

# Question 5

Find all the tracks whose name starts with 'All'.

```python
%%sql
SELECT Name
FROM Track
WHERE Name LIKE "All%"
```

# Question 6

Find all the customer emails that start with "J" and are from gmail.com.

```python
%%sql
SELECT FirstName,LastName,Email
FROM Customer
WHERE Email LIKE "j%gmail.com"
```

# Question 7

Find all the invoices from the billing city Brasília, Edmonton, and Vancouver and sort in descending order by invoice ID.

```python
%%sql
SELECT InvoiceID, BillingCity
FROM Invoice
WHERE BillingCity IN ("Brasília", "Edmonton", "Vancouver")
ORDER BY InvoiceID DESC
```

# Question 8

Show the number of orders placed by each customer (hint: this is found in the invoices table) and sort the result by the number of orders in descending order.

```python
%%sql

SELECT CustomerID, COUNT(CustomerID) AS NumberOfOrders
FROM Invoice
GROUP BY CustomerID
ORDER BY NumberOfOrders DESC
```

# Question 9

Find the albums with 12 or more tracks.

```python
%%sql

SELECT AlbumId, COUNT(AlbumId) AS TrackLength
FROM Track
GROUP BY AlbumId
HAVING COUNT(AlbumId) >= 12
LIMIT 10
```

# Question 10

List each country with the number of distinct customers, the total invoice amount, and the average invoice total. Only include countries where the average invoice total is greater than 5. Sort the results by total invoice amount in descending order.

```python
%%sql

SELECT BillingCountry, COUNT(DISTINCT CustomerID), SUM(Total), AVG(Total)
FROM Invoice
GROUP BY BillingCountry
HAVING AVG(Total) > 5
ORDER BY SUM(Total) DESC
```

# Question 11

From the Invoice table, display BillingCity, total number of invoices, average invoice total, and total revenue. Only show cities where the total revenue is greater than 50. Sort by average invoice total descending.

```python
%%sql

SELECT BillingCity,
       COUNT(*) AS NumberOfInvoices,
       SUM(Total) AS TotalRevenue,
       AVG(Total) AS AvgInvoice
FROM Invoice
GROUP BY BillingCity
HAVING SUM(Total) > 50
ORDER BY AVG(Total) DESC;
```

# Question 12

List each BillingCountry and the number of distinct customers in that country (from the Invoice table). Only include countries with more than 3 customers. Order by number of customers descending.

```python
%%sql

SELECT BillingCountry, COUNT(DISTINCT CustomerID) AS NumberOfCustomers
FROM INVOICE
GROUP BY BillingCountry
HAVING COUNT(DISTINCT CustomerID) > 3
ORDER BY NumberOfCustomers DESC
```
