# Question 1

Using a subquery, find the names of all the tracks for the album "Californication".

```python
%%sql

SELECT Name
FROM Track
WHERE AlbumID = (SELECT AlbumID FROM Album WHERE Title = "Californication")
```

# Question 2

Find the total number of invoices for each customer along with the customer's full name, city and email.

```python
%%sql

SELECT 
    c.FirstName, c.LastName,
    c.City,
    c.Email,
    COUNT(i.InvoiceId) AS TotalInvoices
FROM Customer c
LEFT JOIN Invoice i ON c.CustomerId = i.CustomerId
GROUP BY c.CustomerId, c.FirstName, c.LastName, c.City, c.Email

ORDER BY TotalInvoices DESC, c.LastName, c.FirstName;
```

# Question 3

Retrieve the track name, album, artistID, and trackID for all the albums.

```python
%%sql

SELECT t.Name, a.Title, a.ArtistId, t.TrackID
FROM Album a
LEFT JOIN Track t ON a.AlbumID = t.AlbumID
WHERE a.Title = "For Those About To Rock We Salute You"
```

# Question 4

Retrieve a list with the managers last name, and the last name of the employees who report to him or her.

```python
%%sql

SELECT e1.LastName AS ManagersLastName, e2.LastName AS WhoReportsTo
FROM Employee e1 
LEFT JOIN Employee e2 ON e1.ReportsTo = e2.EmployeeId
```

# Question 5

Find the name and ID of the artists who do not have albums. 

```python
%%sql

SELECT Name, Artist.ArtistID
FROM Artist
LEFT JOIN Album ON Album.ArtistID = Artist.ArtistID
WHERE AlbumID IS NULL
```

# Question 6

Use a UNION to create a list of all the employee's and customer's first names and last names ordered by the last name in descending order.

```python
%%sql

SELECT FirstName, LastName
FROM Customer
UNION
SELECT FirstName, LastName
FROM Employee
ORDER BY LastName DESC
```

# Question 7

See if there are any customers who have a different city listed in their billing city versus their customer city.

```python
%%sql

SELECT Customer.City, Invoice.BillingCity
FROM Customer
LEFT JOIN Invoice ON Customer.CustomerId = Invoice.CustomerId
WHERE Customer.City <> Invoice.BillingCity
```
