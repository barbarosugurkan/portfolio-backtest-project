```python
CtyMpgByCyl = []

for c in cylinders:  # iterate over all the cylinder levels
    summpg = 0
    cyltypecount = 0
    for d in mpg:  # iterate over all dictionaries
        if d['cyl'] == c:  # if the cylinder level type matches,
            summpg += float(d['cty'])  # add the cty mpg
            cyltypecount += 1  # increment the count
    CtyMpgByCyl.append((c, summpg / cyltypecount))  # append the tuple ('cylinder', 'avg mpg')

CtyMpgByCyl.sort(key=lambda x: x[0])
CtyMpgByCyl
```

```python
vehicleclass = set(d["class"] for d in mpg)
vehicleclass
```

```python
HwyMpgByClass = []

for t in vehicleclass:  # iterate over all the vehicle classes
    summpg = 0
    vclasscount = 0
    for d in mpg:  # iterate over all dictionaries
        if d['class'] == t:  # if the cylinder amount type matches,
            summpg += float(d['hwy'])  # add the hwy mpg
            vclasscount += 1  # increment the count
    HwyMpgByClass.append((t, summpg / vclasscount))  # append the tuple ('class', 'avg mpg')

HwyMpgByClass.sort(key=lambda x: x[1])
HwyMpgByClass
```

# **Regex**

```python
import re

# match() başında string olup olmadığına bakar ve boolean verir
# search() herhangi bi yerinde olup olmadığına bakar

text = "this is a good day. But this year was not that good, last year might be better than this year."

if re.search("good", text): #ilk parametre aranacak string ikinci parametre hangi string içinde aranacağı
  print(":)")
else:
  print(":(")


# split() fonksiyonu içine verdiğin ilk parametreyi görünce stringi parçalara böler, ve parçalar o bölünen kelimeyi içermez
re.split("this",text) # -> ['', ' is a good day. But ', ' year was not that good, last year might be better than ', ' year.']

# findall() fonksiyonu ile verilen stringi verilen text içinde kaç defa geçiyosa o kadar içeren bir list alırız
re.findall("this",text) # -> ['this', 'this', 'this']

# aranan stringin verilen string ile başlayıp başlamadığını anlamak için ^ işareti kullanılır (bitiş için de $ işareti kullanılır)
re.search("^this",text) # -> <re.Match object; span=(0, 4), match='this'>
#bize bir re.Match objesi döndü, boolean olarak değeri True'dur. Ayrıca span =(0,4) ile aranan string'in yeri de verilir
if re.search("^this",text):
  print("Text this ile başlıyor")
```

**Patterns and Character Classes**

```python
grades = "ACAABADACB"

#kaç tane B olduğunu bulmak için direkt findall kullanabilirsin
print(re.findall("B",grades))

#kaç tane A veya B olduğunu bulmak için [AB] kullanımı yapman lazım (set operator)
re.findall("[AB]",grades) # -> ['A', 'A', 'A', 'B', 'A', 'A', 'B']

#kaç tane AB veya AC olduğunu bulmak için [A][B-C] kullanımı yapman lazım
re.findall("[A][B-C]",grades) #-> ['AC', 'AB', 'AC']

#aynısını pipe operatorla da yapabilirisn AB|AC şeklinde
re.findall("AB|AC",grades) #-> ['AC', 'AB', 'AC']

#A dışındaki bütün stringler için
re.findall("[^A]",grades) # -> ['C', 'B', 'D', 'C', 'B']

# ^ ifadesi normal "başında bu var mı?" şeklinde bir arama yapıyodu ama set operator (yani []) içine koyunca anlamı değişti
# peki böyle bir kullanım ne sonuç verir?
re.findall("^[^A]",grades) # -> []
#biz burada başında A dışında ne var şeklinde bir sorgulama yaptık, cevap hiçbir şey
re.findall("^[^B]",grades) # -> ['A'] cevabını alırdık, bu da başında B dışında ne var demek
```

**Quantifiers**

```python
grades = "AACAAAADAAABAA"

# Ne kadar üst üste A streaki yaptığını bulmak istiyoruz? {} kullanımı
re.findall("A{2,10}",grades) # -> ['AA', 'AAAA', 'AAA', 'AA'] bizim bburada yazdığımız kod minimum 2 A streaki max 10 A streaki için

re.findall("A{1,1}A{1,1}",grades) # -> ['AA', 'AA', 'AA', 'AA', 'AA']
#bu kod ise iki A karakterinin üst üste kaç kere kullanıldığını veriyor

#önemli bi kısım quantifier hem {m,n} formatında olmalı arada boşluk olmamalı
re.findall("A{2, 2}",grades) #-> []

#hiçbir quantifier yazmazsak varsayılan olarak {1,1} kalır aslında
re.findall("AA",grades) #-> ['AA', 'AA', 'AA', 'AA', 'AA'] iki üsttekiyle aynı

#eğer {} içine sadece tek sayı yazarsan m ve n için o sayıyı yazmışsın gibi anlaşılır
re.findall("A{2}",grades) #-> ['AA', 'AA', 'AA', 'AA', 'AA']

re.findall("A{1,10}B{1,10}",grades) # -> ['AAAB']
#kaç tane en az 1 en fazla 10 A ile en az 1 en fazla B yan yana?

# eğer A* kullanırsan 0 ve daha çok kez anlamı taşır
re.findall("A*",grades) # -> ['AA', '', 'AAAA', '', 'AAA', '', 'AA', '']

# eğer A+ kullanırsan ise 1 veya daha çok anlamı taşır
re.findall("A+",grades) # -> ['AA', 'AAAA', 'AAA', 'AA']

# eğer A? kullanırsan tek tek karakterlere ayırır ve her A'yı alır her A olmayanı da boş döner
re.findall("A?",grades) # -> ['A', 'A', '', 'A', 'A', 'A', 'A', '', 'A', 'A', 'A', '', 'A', 'A', '']
```

```python
with open("/content/ferpa.txt","r") as file:
  wiki = file.read()

#Headerları nasıl alırız?

#Herhangi bir kelimeyle başlayayan (büyük küçük fark etmez) ve sonrasında [edit] ifadesi olan kelimeleri ver
# (buradaki editten önceki \ ifadesi onun set operator amacı dışında direkt string olarak anlaşılması için konmuş)
re.findall("[a-zA-Z]{1,100}\[edit\]",wiki) # -> ['Overview[edit]', 'records[edit]', 'records[edit]']
# Tam istediğimizi alamadık. Sadece headerların son kelimesini aldık

# [\w] ile istediğimiz kelimeyi (harf veya sayı fark etmez) alabiliriz
re.findall("[\w]{1,100}\[edit\]",wiki) # -> ['Overview[edit]', 'records[edit]', 'records[edit]']

# * ile 0 veya daha fazla anlamını verebiliriz
re.findall("[\w]*\[edit\]",wiki) # -> ['Overview[edit]', 'records[edit]', 'records[edit]']

# şimdi [\w ] ile herhangi bir kelime veya boşluğu alabiliriz
re.findall("[\w ]*\[edit\]",wiki) # -> ['Overview[edit]', 'Access to public records[edit]', 'Student medical records[edit]']

# farz edelim tek tek bu başlıkların edit kısmını çıkarmak istiyorum
for title in re.findall("[\w ]*\[edit\]",wiki):
  "print(title[:-6]) -> Overview Access to public records Student medical records"
  # ya da re.split kullanabilliriz (her içine verdiğimizden görünce bi yeni bir string ekliyodu ve bize list veriyodu
  print(re.split("\[",title)[0]) # -> Overview Access to public records Student medical records

print(re.findall("[\w ]*\[edit\]",wiki))
```

**Groups**

```python
import re
re.findall("([\w ]*)(\[edit\])",wiki) # -> [('Overview', '[edit]'), ('Access to public records', '[edit]'), ('Student medical records', '[edit]')]
# () () ifadeleri kullanarak istediklerimizi grupladık

for item in re.finditer("([\w ]*)(\[edit\])",wiki):
  print(item.groups()) # -> ('Overview', '[edit]') ('Access to public records', '[edit]') ('Student medical records', '[edit]')

for item in re.finditer("([\w ]*)(\[edit\])",wiki):
  print(item.group(1)) # -> Overview Access to public records Student medical records

# ?P<isim> şeklinde isim verilebilir
for item in re.finditer("(?P<title>[\w ]*)(?P<edit_link>\[edit\])",wiki):
  print(item.groupdict()) # -> {'title': 'Overview', 'edit_link': '[edit]'} {'title': 'Access to public records', 'edit_link': '[edit]'} {'title': 'Student medical records', 'edit_link': '[edit]'}

# Daha fazlası::
# Ok, we have seen how we can match individual character patterns with [], how we can group matches together
# using (), and how we can use quantifiers such as *, ?, or m{n} to describe patterns. Something I glossed
# over in the previous example was the \w, which standards for any word character. There are a number of
# short hands which are used with regexes for different kinds of characters, including:
# a . for any single character which is not a newline
# a \d for any digit
# and \s for any whitespace character, like spaces and tabs
# There are more, and a full list can be found in the python documentation for regexes
```
