## Soru 1: Ã–ÄŸrenci NotlarÄ±nÄ±n Temizlenmesi ve Analizi

Bir okulun Ã¶ÄŸrenci notlarÄ±nÄ± iÃ§eren "student_grades.csv" adlÄ± bir CSV dosyasÄ± verilmiÅŸ. Dosyada ÅŸu sÃ¼tunlar bulunuyor:

- **Student**: Ã–ÄŸrencinin tam adÄ± (Ã¶r. "Ahmet YÄ±lmaz")
- **Subject**: Ders adÄ±
- **Grade**: Not (bazÄ± giriÅŸler "85", bazÄ±larÄ± "85 puan" gibi tutarsÄ±z formatta)
- **Date**: SÄ±nav tarihi (Ã¶r. "2023-10-15", "15/10/2023" gibi farklÄ± formatlarda)

Ancak veri setinde bazÄ± sorunlar var:
- "Grade" sÃ¼tununda bazÄ± notlar metin iÃ§eriyor (Ã¶r. "85 puan").
- "Date" sÃ¼tununda tarih formatlarÄ± tutarsÄ±z.
- BazÄ± satÄ±rlarda eksik deÄŸerler (None) var.

**GÃ¶revlerin:**
a. Veri setini bir pandas DataFrame'e yÃ¼kle.  
b. "Grade" sÃ¼tununu temizle; metinleri kaldÄ±rÄ±p tÃ¼m deÄŸerleri sayÄ±sal (float) hale getir.  
c. "Date" sÃ¼tununu standart bir datetime formatÄ±na Ã§evir (Ã¶r. "YYYY-MM-DD").  
d. Eksik "Grade" deÄŸerlerini o dersin ortalamasÄ±yla doldur.  
e. Her ders iÃ§in ortalama notu hesapla.  
f. En yÃ¼ksek ortalama nota sahip dersi bul.

````python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/student_grades.csv")

df["Grade"] = df["Grade"].str.extract("([\d]{1,3})(?:\s[\w]*)?")

def fix_date(x):
  if isinstance(x, str):
    x = x.strip()
    if "/" in x:
      return pd.to_datetime(x, dayfirst=True, errors="coerce")
    else:
      return pd.to_datetime(x, errors="coerce")
  return pd.NaT

df["Date"] = df["Date"].apply(fix_date)
df["Date"] = pd.to_datetime(df["Date"])

df["Grade"] = pd.to_numeric(df["Grade"], errors="coerce")

def find_average_and_write(row):
  if pd.isna(row["Grade"]):
    average = int(np.mean(df[df["Subject"]==row["Subject"]]["Grade"].dropna()))
    row["Grade"] = average
    return row
  else:
    return row

df = df.apply(find_average_and_write,axis="columns")

for subject in list(df["Subject"].unique()):
  ortalama = np.mean(df[df["Subject"] == subject]["Grade"])
  print(f"{subject} dersinin ortalamasÄ±: {ortalama}")

df
````

````python
import pandas as pd
import numpy as np
from scipy.stats import ttest_rel
import re

cities=pd.read_html("/content/drive/MyDrive/Colab Notebooks/datasets/Module4 assignments/wikipedia_data.htm")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]
nfl_df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/Module4 assignments/nfl.csv")
mlb_df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/Module4 assignments/mlb.csv")
nba_df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/Module4 assignments/nba.csv")
nhl_df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/Module4 assignments/nhl.csv")

cities["Metropolitan area"] = cities["Metropolitan area"].str.replace(",","").str.replace("â€“"," ")
cities_list = list(cities["Metropolitan area"])
cities = cities.rename(columns={"Population (2016 est.)[8]":"population","Metropolitan area":"city"})

def matching_teams_with_cities(cities_list,team_list):
  match_dict = {}
  for team in team_list:
    for city in cities_list:
      if team.split()[0] == city:
        match_dict[team] = city
      elif " ".join(team.split()[0:2]) == city:
        match_dict[team] = city
      elif "New York" in team:
        match_dict[team] = "New York City"
      elif "Washington" in team:
        match_dict[team] = "Washington D.C."
      elif "Dallas" in team:
        match_dict[team] = "Dallas Fort Worth"
      elif "Tampa Bay" in team:
        match_dict[team] = "Tampa Bay Area"
      elif "Florida" in team:
        match_dict[team] = "Miami Fort Lauderdale"
      elif "Devils" in team:
        match_dict[team] = "New York City"
      elif "Carolina" in team:
        match_dict[team] = "Raleigh"
      elif "Minnesota" in team:
        match_dict[team] = "Minneapolis Saint Paul"
      elif "Colorado" in team:
        match_dict[team] = "Denver"
      elif "Vegas" in team:
        match_dict[team] = "Las Vegas"
      elif "Anaheim" in team:
        match_dict[team] = "Los Angeles"
      elif "San Jose" in team:
        match_dict[team] = "San Francisco Bay Area"
      elif "Arizona" in team:
        match_dict[team] = "Phoenix"
      elif "Indiana" in team:
        match_dict[team] = "Indianapolis"
      elif "Miami" in team:
        match_dict[team] = "Miami Fort Lauderdale"
      elif "Brooklyn" in team:
        match_dict[team] = "New York City"
      elif "Golden State" in team:
        match_dict[team] = "San Francisco Bay Area"
      elif "Utah" in team:
        match_dict[team] = "Salt Lake City"
      elif "San Francisco" in team:
        match_dict[team] = "San Francisco Bay Area"
      elif "Oakland" in team:
        match_dict[team] = "San Francisco Bay Area"
      elif "Tennessee" in team:
        match_dict[team] = "Nashville"
      elif "Patriots" in team:
        match_dict[team] = "Boston"
      elif "Texas" in team:
        match_dict[team] = "Dallas Fort Worth"
  return match_dict

nhl_df["team"] = nhl_df["team"].str.replace("*","")
nhl_df = nhl_df[(nhl_df["GP"] == "82") & (nhl_df["year"] == 2018)]
nhl_df["city"] = nhl_df["team"].map(matching_teams_with_cities(cities_list,list(nhl_df["team"])))
nhl_df = nhl_df.sort_values(by="city")
nhl_df = pd.merge(nhl_df,cities[["city","population"]],how="left",on="city")
nhl_df = nhl_df.set_index(["city","team"])
nhl_df[["W", "L","population"]] = nhl_df[["W", "L","population"]].apply(pd.to_numeric, errors='coerce')
nhl_df["nhl_wl_ratio"] = nhl_df["W"] / (nhl_df["W"] + nhl_df["L"])

nba_df = nba_df[nba_df["year"] == 2018]
nba_df["team"] = nba_df["team"].replace(to_replace="\*?\s\(\d{1,2}\)",value="",regex=True)
nba_df["city"] = nba_df["team"].map(matching_teams_with_cities(cities_list,list(nba_df["team"])))
nba_df = nba_df.sort_values(by="city")
nba_df = pd.merge(nba_df,cities[["city","population"]],how="left",on="city")
nba_df = nba_df.set_index(["city","team"])
nba_df[["W", "L","population"]] = nba_df[["W", "L","population"]].apply(pd.to_numeric, errors='coerce')
nba_df["nba_wl_ratio"] = nba_df["W"] / (nba_df["W"] + nba_df["L"])

mlb_df = mlb_df[mlb_df["year"] == 2018]
mlb_df["city"] = mlb_df["team"].map(matching_teams_with_cities(cities_list,list(mlb_df["team"])))
mlb_df = mlb_df.sort_values(by="city")
mlb_df = pd.merge(mlb_df,cities[["city","population"]],how="left",on="city")
mlb_df = mlb_df.set_index(["city","team"])
mlb_df[["W", "L","population"]] = mlb_df[["W", "L","population"]].apply(pd.to_numeric, errors='coerce')
mlb_df["mlb_wl_ratio"] = mlb_df["W"] / (mlb_df["W"] + mlb_df["L"])

nfl_df = nfl_df[nfl_df["year"] == 2018]
nfl_df = nfl_df[(nfl_df["T"] == "0") | (nfl_df["T"] == "1")]
nfl_df["team"] = nfl_df["team"].str.replace("*","").str.replace("+","")
nfl_df["city"] = nfl_df["team"].map(matching_teams_with_cities(cities_list,list(nfl_df["team"])))
nfl_df = nfl_df.sort_values(by="city")
nfl_df = pd.merge(nfl_df,cities[["city","population"]],how="left",on="city")
nfl_df = nfl_df.set_index(["city","team"])
nfl_df[["W", "L","population"]] = nfl_df[["W", "L","population"]].apply(pd.to_numeric, errors='coerce')
nfl_df["nfl_wl_ratio"] = nfl_df["W"] / (nfl_df["W"] + nfl_df["L"])

nfl_and_mlb_df = pd.merge(nfl_df.groupby("city").agg({"nfl_wl_ratio":"mean"}),mlb_df.groupby("city").agg({"mlb_wl_ratio":"mean"}),how="outer",left_index=True,right_index=True)
nba_nfl_and_mlb_df = pd.merge(nfl_and_mlb_df,nba_df.groupby("city").agg({"nba_wl_ratio":"mean"}),how="outer",left_index=True,right_index=True)
nhl_nba_nfl_and_mlb_df = pd.merge(nba_nfl_and_mlb_df,nhl_df.groupby("city").agg({"nhl_wl_ratio":"mean"}),how="outer",left_index=True,right_index=True)

sports = ['NFL', 'NBA', 'NHL', 'MLB']
p_values = pd.DataFrame({k:np.nan for k in sports}, index=sports)

for sport1 in sports:
  for sport2 in sports:
    if sport1 == sport2:
      p_values.loc[sport1, sport2] = np.nan
    else:
      ratio1 = sport1.lower() + "_wl_ratio"
      ratio2 = sport2.lower() + "_wl_ratio"

      series1 = nhl_nba_nfl_and_mlb_df[ratio1]
      series2 = nhl_nba_nfl_and_mlb_df[ratio2]

      t_statistic, p_value = ttest_rel(series1, series2, nan_policy='omit')

      p_values.loc[sport1, sport2] = p_value

p_values
````
